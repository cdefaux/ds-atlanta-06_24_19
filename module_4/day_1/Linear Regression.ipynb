## Simple linear regression

### Part I. Theory Simple Linear Regression

- An approach for predicting a **quantitative response** Y on the basis of a single predictor variable X. 
- Assumes that there is approximately a linear relationship between X and Y . 

Mathematically: 
- we can write the linear relationship as
${Y_i = β_0 + β_1x_i+ \epsilon_i}$
  -  **Population regression line**, which is the best linear approximation to the true relationship between X and Y.
  - ${\epsilon_i}$ **is a random error**
  - **${\epsilon_i}$ are assumed to independent and normally distributed with mean μ and standard deviation σ** 
- We say that we are regressing Y on X (or Y onto X)

### Interpreting coefficients
- ${β_0}$ and ${β_1}$ are known as the model coefficients or parameters. 
- ${β_0}$ and ${β_1}$ are two unknown constants that represent the intercept and slope terms in the linear model. 
- **${β_0}$ is the intercept term**—the expected value of Y when X = 0
- **${β_1}$ is the slope** (the average increase in Y associated with a one-unit increase in X)

### Model Assumptions

- The mean ${Y_i}$ is a linear function of ${x_i}$
- The ${Y_i}$ have a common variance ${\sigma^2}$, which is the same for all values of x
- The errors ${\epsilon_i}$ ( and hence the observartions ${Y_i}$) are normally distributed
- The errors ${\epsilon_i}$ ( and hence the observartions ${Y_i}$) are independent


##  Line Representing a Simple Linear Regression

- Since we do not have the population measures. We use the form:

$$Y ≈ β_0 + β_1X$$

- We can say that Y “is approximately modeled as”.
- However, this is usually not required. and the line is written:

$$Y = β_0 + β_1X$$

### Predictions
- To obtain estimates of ${\hatβ_0}$ and  ${\hatβ_1}$ of the model parameters ${β_0, β_1}$ we find value of ${β_0, β_1}$ that minimize: 

$${Q = \sum_{i=1}^{n} [y_i-(β_0 + β_1x_i)]^2}$$

- Once we have used our data to produce estimates ${\hatβ_0}$ and  ${\hatβ_1}$, we can predict values of y by using the **least squares (LS) line**:


  $${\hat y = \hat β_0 + \hat β_1x}$$

- ${\hat y}$  indicates a prediction of Y on the basis of X = x ( predicted value of y). 
   - Here we use a hat symbol, ^, to denote the estimated value for an unknown parameter or coefficient, or to denote the predicted value of the response.
- **Note:** ${\hatβ_0}$ and  ${\hatβ_1}$ are the estimates of ${β_0, β_1}$ that will provide the best fit to the data for the linear model

### Residuals
- **used as a foundation to asses the goodness of fit of the LS model.**
- **${e_i=y_i-\hat y_i = y_i-(\hat β_0 + \hat β_1x_i),\ i=1,2,..,n}$**


### Correlation Coefficient (r): 
- numerical measure of strength & direction of linear association.
- unitless quantity that takes values between -1 and 1.

### Correlation vs Causation
- If two variables are correlated this does not necessarily imply that x causes y to change

## Assessing Model Fit

- We focus on deviations from predicted values using residuals terms: 
  - **${e_i=y_i-\hat y_i = y_i-(\hat β_0 + \hat β_1x_i),\ i=1,2,..,n}$**
  
  


#### Examples of residual plots


- (a) is a satisfactory plot with the residuals falling in a horizontal band with no systematic pattern. Such a plot indicates an appropriate regression model. 
- (b) shows residuals falling in a funnel shape. Such a plot indicates increase in variance of residuals and the assumption of constant variance is violated here. Transformation on  may be helpful in this case (see Transformations). 
- If the residuals follow the pattern of (c) or (d), then this is an indication that the linear regression model is not adequate. Addition of higher order terms to the regression model or transformation on  or  may be required in such cases. 
- A plot of residuals may also show a pattern as seen in (e), indicating that the residuals increase (or decrease) as the run order sequence or time progresses. This may be due to factors such as operator-learning or instrument-creep and should be investigated further.

### SSE
- Sum of Square Error (**SSE**)
$$ SSE=\sum_{i=1}^{n} \big( y_{i} - \widehat{y}_{i}  \big)^2 = \sum_{i=1}^{n}e_i^2.$$
### MSE
A useful quantitative measure of a model's performance is the so-called Mean Squared Error (MSE), which is simply the average value of the SSE cost function that we minimize to fit the linear regression model. The MSE is useful to for comparing different regression models or for tuning their parameters via a grid search and cross-validation:

$$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y_{i} - \widehat{y}_{i}  \big)^2.$$

### ${R^2}$ Statistic



- Sometimes it may be more useful to report the coefficient of determination $R^2$, which can be understood as a standardized version of the MSE, for better interpretability of the model performance. 

- ${R^2}$ **statistic provides a measure of fit. It takes the form of a proportion—the proportion of variance explained—and takes on a value between 0 and 1, and is independent of the scale of Y .**
- In other words, $R^2$ is the fraction of response variance that is captured by the model. The $R^2$ value is defined as follows:

$$R^2 = 1 - \frac{SSE}{SST}$$

Here, SST is the total sum of squares $SST = \sum^{n}_{i=1} \big( y_{i} - \bar y \big)^2$, or in other words, it is simply the variance of the response. Let's quickly show that $R^2$ is indeed just the rescaled version of the MSE:

$$R^2 = 1 - \frac{SSE}{SST}$$

$$= 1 - \frac{\frac{1}{n} \sum^{n}_{i=1} \big( y_{i} - \hat{y}_{i}  \big)^2 }{\frac{1}{n} \sum_{i=1}^{n} \big( y_{i} - \bar y \big)^2 }$$
 
$$= 1 - \frac{MSE}{Var(y)}$$

For the training dataset, $R^2$ is bounded between 0 and 1, but it can become negative for the test set. If $R^2$ =1, the model  fits the data perfectly with a
corresponding $MSE = 0$.

## Hypothesis Testing Simple Linear Model

- Since the slope ${\hatβ_1}$ of the true regression line is nonzero when there is a linear relationship, the hypothesis we want to test are 

$${H_0 : There\ is\ no\ relationship\ between\ X\ and\ Y\ vs.\
H_1 : There\ is\ some\ relationship\ between\ X\ and\ Y}$$

- Mathematically, this corresponds to testing
 $${H_0:β_1=0\ vs.\ H_1:β_1 \neq 0}$$
 
 $${t = \frac{\hatβ_1 − 0} {S E (\hatβ_1 )}}$$
 
- Hence, if we see a small p-value, we can infer that there is an association between the predictor and the response. 
- We reject the null hypothesis—that is, we declare a relationship to exist between X and Y —if the p-value is small enough.

### Naive Approach

- Lets try fitting the data using slope intercept formula: $${y=mx+b}$$ 


- We will compare the SSE and MSE we obtain using this model with the **least square estimates**


We observe the intercept $β_0=394.33$ and slope $β_1=\dfrac{150.33-394.33}{36-0}=-6.78$

```python
def point_slope(x):
    b0=394.33
    b1=-6.78
    #b1=-7.625
    estimates =  b0+b1*x
    return estimates
```

```python
point_slope_estimates=point_slope(x)
```

```python
residuals_point_slope=(y-point_slope_estimates)
SSE=sum(np.square(residuals_point_slope))
MSE=SSE/len(x)
print("SSE_point_slope=",SSE,"MSE_point_slope=", MSE)
```

### Using Least Squares Estimates

- SKLEARN 

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
# Note that scikit-learn expects 2D arrays as feature arrays
# so we use `np.newaxis` to append another axis in this case:
x=x[:, np.newaxis]
lr.fit(x, y)
```

```python
# slope
lr.coef_
```

```python
# y-axis intercept
lr.intercept_ 
```

```python
# Make predictions
y_pred = lr.predict(x)
```

```python
# MSE
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y, y_pred)
SSE=MSE*9
print("SSE_Least_Squares=",SSE,"MSE_Least_Squares=", MSE)
```

```python
min_pred = x.min() * lr.coef_ + lr.intercept_
max_pred = x.max() * lr.coef_ + lr.intercept_
plt.scatter(x, y, marker='o')
plt.plot([x.min(), x.max()],
         [min_pred, max_pred],
         color='darkorange',
         linewidth=4)
plt.title('Least Squares Line')
plt.xlabel('x_values')
plt.ylabel('y_values');
```

### Model Diagnostics
- Let's check the four assumptions of our model

### Checking for Linearity
- If regression of y on x is linear, then the plot of $e_i$ vs. $x_i$ should exhibit random scatter around zero

```python
residuals=y-y_pred
```

```python
residuals
```

```python
plt.scatter(x, residuals)
plt.title('Residual Plot')
plt.xlabel('x_values')
plt.ylabel('Residual_Values');
plt.show()
```

- The plot is clearly not linear (it's parabolic) indicating that after filtering out the linear component, there is still a nonlinear component left.
- **Important lesson:** Although ${\hatβ_1}$  is significant it does not necessarily mean the relationship is linear but that there is a significant linear component.

### Checking for Constant Variance
- To chdeck assumption plot $e_i$ vs. $\hat y_i$.
- If the constant variance assumption is correct, the dispersion of the $e_i$'s are approximately constant with respect to $\hat y_i$'s   

```python
plt.scatter(y_pred,residuals)
plt.title('Plot Checking Constant Variance')
plt.xlabel('y_pred')
plt.ylabel('Residual_Values');
plt.show()
```

- plot indicates violation of constant variance assumption

### Checking for Normality
- Residuals should be used to make a normal plot, since they have zero mean and constant variance ( assuming other assumption satisfied) 

```python
from scipy import stats
res = stats.probplot(residuals, plot=plt)
```

- plot exhibits small departure from normality, given small sample size cannot draw a meaningful conclusion. 
- If normality assumption is not satisfied the usual corrective measure is to use transformation on $\hat y_i$'s

### Checking for Independence
- This assumption is violated for time series data due to autocorrelation introduced by time order, called serial or autocorrelation.
- Use Durbin- Watson test.
- The null hypothesis of the test is that there is no serial correlation.
- test statistic is:

$$d = {\sum_{t=2}^T (e_t - e_{t-1})^2 \over {\sum_{t=1}^T e_t^2}}$$  





- d always lies between 0 and 4
- d = 2 indicates no autocorrelation. 
- If d < 2 successive error terms are positively correlated. 
- If d > 2, successive error terms are negatively correlated. 

```python
from statsmodels.stats.stattools import durbin_watson
durbin_watson(residuals)
```
