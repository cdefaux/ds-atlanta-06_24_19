```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('pets.db')
c = conn.cursor()
```

```python
c.execute("""select c.Name, d.Name from cats c
            inner join dogs d on  
            c.Owner_ID = d.Owner_ID
            left join owners on 
            c.Owner_ID = owners.ID
            where owners.Name = 'Avi Flombaum'""")

df = pd.DataFrame(c.fetchall())
df.columns = [i[0] for i in c.description]
df.head(10)
```

```python
c.execute("""select * from cats 
            outer join dogs using(House_Number)
            """)
df = pd.DataFrame(c.fetchall())
df.columns = [i[0] for i in c.description]
df.head(10)
```

```python
c.execute("""select owners.Name, owners.Phone from cats
            join dogs using(Owner_ID)
            join owners on
            owners.ID = cats.Owner_ID & dogs.Owner_ID
            where owners.Address = 'None' 
            and cats.House_Number = '4' 
            and dogs.House_Number = '4'
            """)
df = pd.DataFrame(c.fetchall())
df.columns = [i[0] for i in c.description]
df.head(10)
```
